```python
import os
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json  # 리스트 데이터를 문자열로 변환하여 저장

# ✅ 한글 폰트 설정 (Windows & Mac 대응)
plt.rcParams["font.family"] = "Malgun Gothic"  # Windows (맑은 고딕)
plt.rcParams["axes.unicode_minus"] = False  # 마이너스 기호 깨짐 방지

# ✅ SQLite 데이터베이스 연결
db_path = "asia (1).db"
conn = sqlite3.connect(db_path)
cursor = conn.cursor()

# ✅ `career` 컬럼 초기화 (모든 데이터 삭제)
cursor.execute("UPDATE duty_analysis SET career = NULL;")
conn.commit()
print("🔄 기존 `career` 컬럼 데이터 초기화 완료.")

# ✅ 이미지 저장 폴더 생성
output_dir = "Career_charts"
os.makedirs(output_dir, exist_ok=True)

# ✅ SQL 쿼리 실행: 직무별 경력 요구 비율 조회
sql_query = """
WITH career_data AS (
    SELECT 'PM' AS category, career FROM processing WHERE duty LIKE '%PM%'
    UNION ALL
    SELECT '데이터 직무' AS category, career FROM processing WHERE duty LIKE '%데이터 직무%'
    UNION ALL
    SELECT '백엔드' AS category, career FROM processing WHERE duty LIKE '%백엔드%'
    UNION ALL
    SELECT '인프라 엔지니어' AS category, career FROM processing WHERE duty LIKE '%인프라 엔지니어%'
    UNION ALL
    SELECT '앱 개발자' AS category, career FROM processing WHERE duty LIKE '%앱 개발자%'
    UNION ALL
    SELECT '게임' AS category, career FROM processing WHERE duty LIKE '%게임%'
    UNION ALL
    SELECT 'AI' AS category, career FROM processing WHERE duty LIKE '%AI%'
    UNION ALL
    SELECT '임베디드' AS category, career FROM processing WHERE duty LIKE '%임베디드%'
    UNION ALL
    SELECT '프론트 엔드' AS category, career FROM processing WHERE duty LIKE '%프론트 엔드%'
    UNION ALL
    SELECT 'QA' AS category, career FROM processing WHERE duty LIKE '%QA%'
    UNION ALL
    SELECT '데이터 분석' AS category, career FROM processing WHERE duty LIKE '%데이터 분석%'
    UNION ALL
    SELECT 'VR' AS category, career FROM processing WHERE duty LIKE '%VR%'
    UNION ALL
    SELECT '시스템' AS category, career FROM processing WHERE duty LIKE '%시스템%'
    UNION ALL
    SELECT '블록체인' AS category, career FROM processing WHERE duty LIKE '%블록체인%'
    UNION ALL
    SELECT 'ERP' AS category, career FROM processing WHERE duty LIKE '%ERP%'
    UNION ALL
    SELECT '언어별 개발자' AS category, career FROM processing WHERE duty LIKE '%언어별 개발자%'
)

SELECT
    category AS 직무,
    career AS 연차,
    COUNT(*) AS 채용_공고_개수
FROM career_data
GROUP BY category, career
ORDER BY category, career;
"""

# ✅ 데이터 로드
df = pd.read_sql_query(sql_query, conn)
df.columns = df.columns.str.strip()  # 한글 깨짐 방지
cursor.execute(sql_query)
rows = cursor.fetchall()
print(rows)


# ✅ 연차 컬럼 변환: "0" → "0년차", "1" → "1년차" 형식 적용
df["연차"] = df["연차"].astype(str) + "년차"

# ✅ 직무별 시각화 및 이미지 저장
career_data = {}  # 직무별 이미지 리스트 저장 딕셔너리
unique_jobs = df["직무"].unique()

for job in unique_jobs:
    df_job = df[df["직무"] == job]  # 특정 직무 데이터 필터링
    colors = sns.color_palette("husl", len(df_job["연차"].unique()))  # 연차별 색상 지정

    plt.figure(figsize=(10, 6))
    sns.barplot(data=df_job, x="연차", y="채용_공고_개수", hue="연차", palette=colors, dodge=False, legend=False)

    # ✅ 그래프 꾸미기
    plt.xlabel("경력 연차")
    plt.ylabel("채용 공고 개수")
    plt.title(f"{job} 직무의 연차별 채용 공고 개수")
    plt.xticks(rotation=0)  # 연차 숫자 정자로 정렬
    plt.grid(axis="y", linestyle="--", alpha=0.7)

    # ✅ 이미지 저장 (파일명: "{직무}.png")
    image_path = os.path.join(output_dir, f"{job}.png")
    plt.savefig(image_path, bbox_inches="tight")
    plt.close()

    # ✅ career 데이터 리스트 생성
    summary_text = f"{job} 직군의 채용 공고는 {len(df_job)}개이며, 연차별 채용 분포를 확인할 수 있습니다."
    career_data[job] = json.dumps([image_path, summary_text])

# ✅ 데이터베이스 업데이트 (career 컬럼에 리스트 형태 저장)
for duty, career_json in career_data.items():
    cursor.execute(
        "UPDATE duty_analysis SET career = ? WHERE duty = ?;",
        (career_json, duty)
    )
    conn.commit()

print("✅ 모든 직무별 시각화 이미지가 리스트 형식으로 저장되고 `duty_analysis` 테이블에 반영되었습니다.")

# ✅ DB 연결 종료
conn.close()
```

```python
import os
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json  # 리스트 데이터를 문자열로 변환하여 저장
import openai  # GPT-4 API 호출

# ✅ OpenAI API 키 설정 (본인의 키로 변경 필요)
openai.api_key = ""

# ✅ 한글 폰트 설정 (Windows & Mac 대응)
plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False

# ✅ SQLite 데이터베이스 연결
db_path = "asia (1).db"
conn = sqlite3.connect(db_path)
cursor = conn.cursor()

# ✅ 이미지 저장 폴더 생성
output_dir = "Career_charts"
os.makedirs(output_dir, exist_ok=True)

# ✅ SQL 쿼리 실행: 직무별 경력 요구 비율 조회
sql_query = """
WITH career_data AS (
    SELECT 'PM' AS category, career FROM processing WHERE duty LIKE '%PM%'
    UNION ALL
    SELECT '데이터 직무' AS category, career FROM processing WHERE duty LIKE '%데이터 직무%'
    UNION ALL
    SELECT '백엔드' AS category, career FROM processing WHERE duty LIKE '%백엔드%'
    UNION ALL
    SELECT '인프라 엔지니어' AS category, career FROM processing WHERE duty LIKE '%인프라 엔지니어%'
    UNION ALL
    SELECT '앱 개발자' AS category, career FROM processing WHERE duty LIKE '%앱 개발자%'
    UNION ALL
    SELECT '게임' AS category, career FROM processing WHERE duty LIKE '%게임%'
    UNION ALL
    SELECT 'AI' AS category, career FROM processing WHERE duty LIKE '%AI%'
    UNION ALL
    SELECT '임베디드' AS category, career FROM processing WHERE duty LIKE '%임베디드%'
    UNION ALL
    SELECT '프론트 엔드' AS category, career FROM processing WHERE duty LIKE '%프론트 엔드%'
    UNION ALL
    SELECT 'QA' AS category, career FROM processing WHERE duty LIKE '%QA%'
    UNION ALL
    SELECT '데이터 분석' AS category, career FROM processing WHERE duty LIKE '%데이터 분석%'
    UNION ALL
    SELECT 'VR' AS category, career FROM processing WHERE duty LIKE '%VR%'
    UNION ALL
    SELECT '시스템' AS category, career FROM processing WHERE duty LIKE '%시스템%'
    UNION ALL
    SELECT '블록체인' AS category, career FROM processing WHERE duty LIKE '%블록체인%'
    UNION ALL
    SELECT 'ERP' AS category, career FROM processing WHERE duty LIKE '%ERP%'
    UNION ALL
    SELECT '언어별 개발자' AS category, career FROM processing WHERE duty LIKE '%언어별 개발자%'
)

SELECT
    category AS 직무,
    career AS 연차,
    COUNT(*) AS 채용_공고_개수
FROM career_data
GROUP BY category, career
ORDER BY category, career;
"""

# ✅ 데이터 로드
df = pd.read_sql_query(sql_query, conn)
df.columns = df.columns.str.strip()  # 한글 깨짐 방지
cursor.execute(sql_query)
rows = cursor.fetchall()

# ✅ 데이터 가공: 직군별 연차 데이터 정리
job_data = {}

for duty, career, count in rows:
    if duty not in job_data:
        job_data[duty] = []
    job_data[duty].append((career, count))

# ✅ 직무별 연차별 시각화 + GPT-4 요약
for duty, details in job_data.items():
    # ✅ 1. 채용 공고 수 기준으로 정렬 (내림차순)
    sorted_details = sorted(details, key=lambda x: x[1], reverse=True)

    # ✅ 2. 1순위, 2순위 추출
    first_priority = sorted_details[0] if sorted_details else None
    second_priority = sorted_details[1] if len(sorted_details) > 1 else None

    # ✅ 3. 경력 하한선 (최소 요구 연차)
    min_experience = min(details, key=lambda x: x[0])[0] if details else None

    # ✅ 4. GPT-4 프롬프트 템플릿
    prompt_text = f"""
    "{duty}" 직무의 경력별 채용 현황을 분석해주세요.
    {chr(10).join([f"- {c[0]}년차: {c[1]}건" for c in sorted(details, key=lambda x: x[0])])}

    위 데이터를 기반으로 다음을 분석해주세요:
    1. 가장 많이 요청하는 경력 연차 (1순위): {first_priority[0]}년차 ({first_priority[1]}건) 
    2. 두 번째로 많이 요청하는 경력 연차 (2순위): {second_priority[0]}년차 ({second_priority[1]}건) 
    3. 해당 직무의 경력 하한선 (최소 요구 연차): {min_experience}년차
    """

    # ✅ 5. GPT-4 API 호출
    try:
        response = openai.ChatCompletion.create(
            model="gpt-4",
            messages=[{"role": "user", "content": prompt_text}]
        )
        summary = response["choices"][0]["message"]["content"].strip()
    except Exception as e:
        summary = f"❌ GPT-4 요약 실패: {e}"

    # ✅ 6. 기존 데이터 가져와서 이미지 경로 유지
    cursor.execute("SELECT career FROM duty_analysis WHERE duty = ?;", (duty,))
    existing_data = cursor.fetchone()

    if existing_data and existing_data[0]:  # 기존 데이터가 있을 경우
        try:
            existing_list = json.loads(existing_data[0])  # 기존 데이터를 리스트로 변환
            image_path = existing_list[0]  # 기존 이미지 경로 유지
        except:
            image_path = os.path.join(output_dir, f"{duty}.png")  # 오류 발생 시 기본 이미지 경로
    else:
        image_path = os.path.join(output_dir, f"{duty}.png")  # 기본 이미지 경로

    # ✅ 7. 최종 저장할 리스트 생성
    career_data = json.dumps([image_path, summary], ensure_ascii=False)

    # ✅ 8. DB 업데이트
    cursor.execute(
        "UPDATE duty_analysis SET career = ? WHERE duty = ?;",
        (career_data, duty)
    )
    conn.commit()

    print(f"✅ {duty} 직무 요약 완료 및 저장: {career_data}")

# ✅ DB 연결 종료
conn.close()
print("🎯 모든 직무별 연차 분석이 완료되었습니다.")
```

```python
import openai
import sqlite3
import json  # 리스트 데이터를 문자열로 변환하여 저장

# ✅ OpenAI API 키 설정
openai.api_key = ""

# ✅ SQLite 데이터베이스 연결
db_path = "asia (1).db"
conn = sqlite3.connect(db_path)
cursor = conn.cursor()

# ✅ 직무별 연차 데이터 조회
cursor.execute("""
    SELECT duty, career FROM duty_analysis;
""")
rows = cursor.fetchall()

# ✅ 각 직무별 연차 데이터 조회
cursor.execute("""
    SELECT duty, career, COUNT(*) as job_count
    FROM processing
    GROUP BY duty, career
    ORDER BY duty, career;
""")
career_data = cursor.fetchall()

# ✅ 데이터 변환 (직무별 연차별 데이터 정리)
job_career_dict = {}
for duty, career, job_count in career_data:
    if duty not in job_career_dict:
        job_career_dict[duty] = []
    job_career_dict[duty].append((career, job_count))

# ✅ 직무별 GPT-4 요약 생성 및 업데이트
for duty, career in rows:
    if duty not in job_career_dict:
        continue  # 해당 직무에 대한 데이터가 없으면 건너뜀

    # ✅ GPT-4 프롬프트 생성
    job_details = "\n".join([f"- {c[0]}년차: {c[1]}건" for c in job_career_dict[duty]])
    prompt = f"""
    "{duty}" 직무의 경력별 채용 현황을 분석해주세요.
    {job_details}

    위 데이터를 기반으로 다음을 분석해주세요:
    1. 가장 많이 요청하는 경력 연차 (1순위) 및 해당 연차의 채용 공고 수
    2. 두 번째로 많이 요청하는 경력 연차 (2순위) 및 해당 연차의 채용 공고 수
    3. 해당 직무의 경력 하한선 (최소 요구 연차)

    최종 분석 내용을 간결하게 요약해서 제공해주세요.
    """

    try:
        # ✅ GPT-4 API 호출
        response = openai.chat.completions.create(
            model="gpt-4",
            messages=[{"role": "user", "content": prompt}]
        )
        gpt4_summary = response.choices[0].message.content.strip()

        # ✅ 기존 career 칼럼의 값 가져오기 (이미지 경로 유지)
        cursor.execute("SELECT career FROM duty_analysis WHERE duty = ?;", (duty,))
        existing_career_data = cursor.fetchone()

        if existing_career_data and existing_career_data[0]:
            parts = existing_career_data[0].split(")", 1)  # ")"을 기준으로 이미지 경로와 기존 요약 분리
            image_path = parts[0] + ")" if len(parts) > 1 else existing_career_data[0]  # 이미지 경로 유지
        else:
            image_path = ""

        # ✅ `career` 값 업데이트 (리스트 형식으로 저장)
        career_list = [image_path, gpt4_summary]
        career_json = json.dumps(career_list, ensure_ascii=False)  # JSON 변환 (한글 깨짐 방지)

        cursor.execute(
            "UPDATE duty_analysis SET career = ? WHERE duty = ?;",
            (career_json, duty)
        )
        conn.commit()

        print(f"✅ {duty} 요약 완료 (GPT-4 적용): {gpt4_summary}")

    except Exception as e:
        print(f"❌ {duty} 요약 실패: {e}")

# ✅ DB 연결 종료
conn.close()
print("🎯 모든 직무의 경력 요구 사항 요약이 GPT-4로 업데이트되었습니다.")
```
