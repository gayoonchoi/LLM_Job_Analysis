```python
# 데이터 준비
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

# 샘플 데이터: 기술 조합 트랜잭션
data = {
    'Python': [1, 1, 0, 1, 0],
    'Java': [1, 0, 1, 1, 0],
    'JavaScript': [0, 1, 1, 1, 1],
    'React': [0, 1, 0, 1, 1],
    'SQL': [1, 1, 1, 0, 1],
}
df = pd.DataFrame(data)

# 데이터 변환: 1/0 -> True/False
df = df.astype(bool)

# 1. 빈번한 항목 집합 생성
frequent_itemsets = apriori(df, min_support=0.5, use_colnames=True)

print("Frequent Itemsets:")
print(frequent_itemsets)
```

```python
# 2. 연관성 규칙 생성
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)

print("\nAssociation Rules:")
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```

```python
import pyfpgrowth

# 데이터 준비
transactions = [
    ['Python', 'Java', 'SQL'],
    ['JavaScript', 'React', 'SQL'],
    ['Java', 'JavaScript', 'SQL'],
    ['Python', 'Java', 'JavaScript'],
    ['React', 'SQL']
]

import pyfpgrowth


# 1. 패턴 찾기
# 최소 발생 횟수를 지정 (support_threshold)
patterns = pyfpgrowth.find_frequent_patterns(transactions, support_threshold=2)
print("Frequent Patterns:", patterns)

# 2. 연관 규칙 생성
rules = pyfpgrowth.generate_association_rules(patterns, confidence_threshold=0.7)
print("Association Rules:", rules)
```

```python
# 필요한 라이브러리 설치

import pandas as pd
import json
from transformers import pipeline

# CSV 파일 로드
file_path = "/content/2025-01-21-12_final.csv"
data = pd.read_csv(file_path)

# json 컬럼 추출
if "json" not in data.columns:
    raise ValueError("CSV 파일에 'json' 컬럼이 존재하지 않습니다.")

json_data = data['json'][:10]  # 상위 10개 데이터만 사용

# 요약 모델 로드
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# 결과 저장 리스트
results = []

# 데이터 처리
for idx, entry in enumerate(json_data):
    try:
        # JSON 파싱
        job_info = json.loads(entry)

        # 회사 정보 추출
        company = job_info.get("jobPosition", {}).get("company", {})
        company_name = company.get("name", "정보 없음")
        employees_count = company.get("employeesCount", "정보 없음")
        address = company.get("address", "정보 없음")

        # 채용 및 전형 절차 정보
        hiring_process = job_info.get("jobPosition", {}).get("description", "정보 없음")

        # 텍스트 결합
        raw_text = f"""
        회사 이름: {company_name}
        회사 규모: {employees_count}명
        회사 위치: {address}
        채용 및 전형 절차: {hiring_process}
        """

        # 요약 수행
        summary = summarizer(raw_text, max_length=100, min_length=30, do_sample=False)[0]["summary_text"]

        # 결과 저장
        results.append({
            "Company Name": company_name,
            "Employee Counts": employees_count,
            "Address": address,
            "Hiring Process": hiring_process,
            "Summary": summary,
        })

    except Exception as e:
        print(f"Error processing entry {idx}: {e}")

# 결과를 데이터프레임으로 변환
result_df = pd.DataFrame(results)

# 데이터프레임 출력
print(result_df.head())

# 결과를 CSV로 저장
output_path = "/content/job_summary_results.csv"
result_df.to_csv(output_path, index=False)
print(f"Results saved to {output_path}")
```
