```python
import pandas as pd
import subprocess

# 1. 크롤링 데이터 로드
# 크롤링한 데이터가 'crawled_data.csv' 파일에 저장되어 있다고 가정합니다.
# 파일 경로가 다를 경우 경로를 수정해주세요.
crawled_data_path = '/content/techstack_data (1).csv'  # 예: '/path/to/crawled_data.csv'

try:
    df = pd.read_csv(crawled_data_path)
    print("크롤링 데이터 로드 성공")
except FileNotFoundError:
    print(f"파일을 찾을 수 없습니다: {crawled_data_path}")
    # 필요 시 파일을 불러오는 추가 코드를 작성하세요.
    raise
```

```python
# 2. 사용자 사전 생성
# 'name' 컬럼에서 고유한 단어를 추출합니다.
if 'name' not in df.columns:
    raise ValueError("'name' 컬럼이 데이터에 존재하지 않습니다.")

unique_names = df['name'].dropna().unique()
print(f"추출된 고유 단어 수: {len(unique_names)}")

# MeCab 사용자 사전 형식에 맞게 데이터 포맷팅
# 형식: surface, reading, base_form, POS1, POS2, POS3, conjugated_form, inflection_type, inflection_form, additional_info
# 여기서는 간단히 '*'로 채웁니다.
user_dict_entries = [
    f"{name},{name},{name},NNP,*,*,{name},*,*,*" for name in unique_names
]

# 사용자 사전 CSV 파일 경로
user_dict_csv_path = '/content/user_dict.csv'  # 원하는 경로로 수정 가능

# CSV 파일로 저장
with open(user_dict_csv_path, 'w', encoding='utf-8') as f:
    f.write('\n'.join(user_dict_entries))

print(f"사용자 사전 CSV 파일 생성 완료: {user_dict_csv_path}")

```

# 사용자 사전 문제는 나중에 해결하고 전체 데이터에서 형태소 분석해보기

```python
import pandas as pd

# CSV 파일 경로
file_path = '/content/merged_df.csv'

# CSV 파일 읽기
df = pd.read_csv(file_path)

# 데이터 확인
print("원본 데이터:")
print(df.head())
```

```python
import re
# MeCab을 사용하여 텍스트에서 명사를 추출하는 함수를 정의합니다.
def analyze_morph(text):
    parsed = mecab.parse(text)
    morphs = []
    for line in parsed.split('\n'):
        if line == 'EOS' or line == '':
            break
        parts = line.split('\t')
        if len(parts) > 1:
            word = parts[0]
            feature = parts[3].split('-')  # 품사 정보는 4번째 필드에 위치
            pos = feature[0]  # 품사 정보
            morphs.append((word, pos))
    return morphs
```

```python
# 명사와 영어 단어를 추출하는 함수를 정의합니다.
def extract_nouns_and_english(morph_result):
    nouns = []
    english = []
    for word, pos in morph_result:
        if pos.startswith('NN'):  # 명사 추출
            nouns.append(word)
        if re.match(r'^[A-Za-z]+$', word):  # 영어 단어 추출
            english.append(word)
    return pd.Series({'nouns': nouns, 'english_words': english})
```

```python
# 필요한 열 선택 (예: title, description, requirement, preferredExperience)
text_columns = ['title', 'description', 'requirement', 'preferredExperience']

# 텍스트 컬럼 결합 (분석 대상 텍스트 생성)
df['combined_text'] = df[text_columns].fillna('').agg(' '.join, axis=1)

# 형태소 분석 적용
df['morph_analysis'] = df['combined_text'].apply(analyze_morph)

# 명사 및 영어 단어 추출 적용
df[['nouns', 'english_words']] = df['morph_analysis'].apply(extract_nouns_and_english)
```
