```python
import pandas as pd
from transformers import pipeline
from sklearn.feature_extraction.text import CountVectorizer

# Hugging Face의 Summarization 파이프라인 로드
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# CSV 파일 경로
csv_file_path = '/content/test_df.csv'  # 업로드된 파일 경로

# CSV 파일 읽기
def load_csv(file_path):
    return pd.read_csv(file_path)

# 텍스트 요약하는 함수
def summarize_text(text):
    if not isinstance(text, str) or not text.strip():  # 텍스트가 비어있거나 문자열이 아닐 경우
        return "내용이 없습니다."
    summary = summarizer(text, max_length=150, min_length=50, do_sample=False)
    return summary[0]['summary_text']

# 키워드로 관련 텍스트 찾기
def search_for_answer(df, query):
    vectorizer = CountVectorizer(stop_words='english')
    X = vectorizer.fit_transform(df['preferredExperience'].apply(str))  # 'preferredExperience' 컬럼을 텍스트로 변환
    query_vector = vectorizer.transform([query])  # 질문을 벡터화
    similarities = (X * query_vector.T).toarray()  # 질문과 각 텍스트 간의 유사도 계산
    return similarities

# 가장 유사한 텍스트를 선택하고 요약
def get_top_answer(df, query, top_n=3):
    similarities = search_for_answer(df, query)
    top_indexes = similarities.flatten().argsort()[-top_n:][::-1]  # 가장 유사한 상위 n개 텍스트 인덱스 추출
    answers = []
    for index in top_indexes:
        text = df.iloc[index]['preferredExperience']
        summary = summarize_text(text)
        answers.append(summary)
    return answers

# CSV 파일을 읽고 요약하는 작업 수행
def main():
    # CSV 파일 읽기
    df = load_csv(csv_file_path)

    # 질문: 프론트엔드, 백엔드, 풀스택
    queries = ['프론트엔드', '백엔드', '풀스택']

    for query in queries:
        print(f"### '{query}' 관련 텍스트 요약")

        # 관련 텍스트를 추출하고 요약
        summaries = get_top_answer(df, query)

        for idx, summary in enumerate(summaries, 1):
            print(f"**요약 {idx}:**")
            print(summary)
            print("-" * 50)

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
from transformers import pipeline

# Hugging Face의 Summarization 파이프라인 로드
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# CSV 파일 경로
csv_file_path = '/content/test_df.csv'  # 업로드된 파일 경로

# CSV 파일 읽기
def load_csv(file_path):
    return pd.read_csv(file_path)

# 텍스트 요약하는 함수 (긴 텍스트를 여러 부분으로 나누어 요약)
def summarize_text(text):
    max_length = 1024  # 모델이 처리할 수 있는 최대 텍스트 길이
    step = max_length // 2  # 각 부분에서 겹치는 부분을 두어 잘리도록 함

    # 텍스트를 여러 부분으로 나누기
    text_parts = [text[i:i + max_length] for i in range(0, len(text), max_length - step)]

    # 각 부분을 요약하고 합치기
    summaries = [summarizer(part, max_length=150, min_length=50, do_sample=False)[0]['summary_text'] for part in text_parts]

    # 모든 요약을 합쳐서 최종 요약을 생성
    final_summary = ' '.join(summaries)
    return final_summary

# 키워드로 관련 텍스트 찾기
def find_text_by_keyword(df, keyword):
    matching_rows = df[df.apply(lambda row: row.astype(str).str.contains(keyword, case=False).any(), axis=1)]
    return matching_rows

# CSV 파일을 읽고 요약하는 작업 수행
def main():
    # CSV 파일 읽기
    df = load_csv(csv_file_path)

    # 키워드 목록
    keywords = ['프론트엔드', '백엔드', '풀스택']

    for keyword in keywords:
        print(f"### '{keyword}' 관련 텍스트 요약")

        # 키워드와 관련된 텍스트 찾기
        matching_rows = find_text_by_keyword(df, keyword)

        if not matching_rows.empty:
            for index, row in matching_rows.iterrows():
                text = row['preferredExperience']  # 해당 행의 'preferredExperience' 컬럼에서 텍스트 추출
                summary = summarize_text(text)
                print(f"**원본 텍스트:** {text}")
                print(f"**요약:** {summary}")
                print("-" * 50)
        else:
            print(f"'{keyword}' 관련 텍스트가 없습니다.")
            print("-" * 50)

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
from transformers import pipeline

# Hugging Face의 Summarization 파이프라인 로드
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# CSV 파일 경로
csv_file_path = '/content/test_df.csv'  # 업로드된 파일 경로

# CSV 파일 읽기
def load_csv(file_path):
    return pd.read_csv(file_path)

# 텍스트 요약하는 함수
def summarize_text(text):
    if not isinstance(text, str) or not text.strip():  # 텍스트가 비어있거나 문자열이 아닐 경우
        return "내용이 없습니다."
    summary = summarizer(text, max_length=150, min_length=50, do_sample=False)
    return summary[0]['summary_text']

# 텍스트 청크로 나누기
def chunk_text(text, chunk_size=512):
    # 텍스트가 문자열인지 확인
    if not isinstance(text, str):
        text = str(text)  # 숫자나 NaN 값이 있을 경우 문자열로 변환

    # 텍스트를 청크 크기(chunk_size)로 나눕니다.
    chunks = [text[i:i + chunk_size] for i in range(0, len(text), chunk_size)]
    return chunks

# FAISS 인덱스 생성 및 검색
def create_faiss_index(texts):
    model = SentenceTransformer('paraphrase-MiniLM-L6-v2')  # 텍스트를 임베딩할 모델
    embeddings = model.encode(texts)  # 텍스트 임베딩 생성

    # FAISS 인덱스 생성
    dim = embeddings.shape[1]
    index = faiss.IndexFlatL2(dim)  # L2 거리 측정 방법을 사용
    index.add(np.array(embeddings))  # 임베딩 추가
    return index, embeddings

# 검색어에 해당하는 텍스트 찾기
def search_faiss(index, query, k=3):
    model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
    query_embedding = model.encode([query])  # 검색어 임베딩

    # FAISS에서 가장 가까운 k개의 벡터 찾기
    D, I = index.search(np.array(query_embedding), k)
    return I  # 해당하는 인덱스 반환

# CSV 파일을 읽고 요약하는 작업 수행
def main():
    # CSV 파일 읽기
    df = load_csv(csv_file_path)

    # 1. 텍스트 청크 나누기
    texts = []
    for index, row in df.iterrows():
        text = row['preferredExperience']  # 해당 행의 'preferredExperience' 컬럼에서 텍스트 추출
        if pd.isna(text):  # NaN 값은 건너뜁니다
            continue
        chunks = chunk_text(text)  # 텍스트를 청크로 나눕니다.
        texts.extend(chunks)  # 모든 청크를 텍스트 목록에 추가

    # 2. FAISS 인덱스 생성
    index, embeddings = create_faiss_index(texts)

    # 3. 검색어를 기반으로 텍스트 찾기
    queries = ['프론트엔드', '백엔드', '풀스택']  # 검색어 목록

    for query in queries:
        print(f"### '{query}' 관련 텍스트 요약")

        # 검색하여 관련 텍스트 인덱스를 찾습니다.
        indexes = search_faiss(index, query)

        # 검색된 텍스트 요약
        for idx in indexes[0]:  # top-k 텍스트에 대해 요약
            summary = summarize_text(texts[idx])
            print(f"**요약:** {summary}")
            print("-" * 50)

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
from sentence_transformers import SentenceTransformer
import pinecone
import numpy as np
from transformers import pipeline
import os

# Hugging Face의 Summarization 파이프라인 로드
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Pinecone API 키와 환경 설정 - job rag
PINECONE_API_KEY = 'pcsk_6pCKjq_La4aZBqiTUciaVC8mvByb6NhbmMdQEVfyJFmaBaRakBVQMrF4GETud3saMK6Pif'
PINECONE_ENVIRONMENT = 'eu-central1-gcp'  # Pinecone 환경 (예: 'us-east1-gcp')

# Pinecone 클라이언트 초기화
# pinecone.init(api_key=PINECONE_API_KEY, environment=PINECONE_ENVIRONMENT)


# Pinecone 클라이언트 초기화 (새로운 방식)
from pinecone import Pinecone, ServerlessSpec

# Pinecone 클라이언트 인스턴스 생성
pc = Pinecone(api_key=PINECONE_API_KEY)

# CSV 파일 경로
csv_file_path = '/content/test_df.csv'  # 업로드된 파일 경로

# CSV 파일 읽기
def load_csv(file_path):
    return pd.read_csv(file_path)

# 텍스트 요약하는 함수
def summarize_text(text):
    if not isinstance(text, str) or not text.strip():  # 텍스트가 비어있거나 문자열이 아닐 경우
        return "내용이 없습니다."
    summary = summarizer(text, max_length=150, min_length=50, do_sample=False)
    return summary[0]['summary_text']

# 텍스트 청크로 나누기
def chunk_text(text, chunk_size=512):
    # 텍스트가 문자열인지 확인
    if not isinstance(text, str):
        text = str(text)  # 숫자나 NaN 값이 있을 경우 문자열로 변환

    # 텍스트를 청크 크기(chunk_size)로 나눕니다.
    chunks = [text[i:i + chunk_size] for i in range(0, len(text), chunk_size)]
    return chunks

# Pinecone DB에 텍스트 임베딩 추가
def add_to_pinecone(texts):
    model = SentenceTransformer('paraphrase-MiniLM-L6-v2')  # 텍스트를 임베딩할 모델
    embeddings = model.encode(texts)  # 텍스트 임베딩

    # Pinecone에 임베딩 추가
    index_name = "text-embeddings"

    # 인덱스가 존재하지 않으면 새로 생성
    if index_name not in pc.list_indexes().names():
        pc.create_index(
            name=index_name,
            dimension=embeddings.shape[1],
            metric='cosine',
            spec=ServerlessSpec(cloud='aws', region='us-west-2')
        )

    index = pc.index(index_name)

    # 각 텍스트에 대해 임베딩을 Pinecone DB에 저장
    ids = [str(i) for i in range(len(texts))]
    vectors = [{"id": ids[i], "values": embeddings[i].tolist(), "metadata": {"text": texts[i]}} for i in range(len(texts))]

    # 데이터 추가
    index.upsert(vectors=vectors)

# Pinecone에서 쿼리로 텍스트 찾기
def search_pinecone(query, k=3):
    model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
    query_embedding = model.encode([query])  # 쿼리 임베딩

    # Pinecone에서 검색
    index_name = "text-embeddings"
    index = pc.index(index_name)

    # 검색
    results = index.query(query_embedding[0].tolist(), top_k=k, include_metadata=True)

    # 검색된 텍스트 반환
    return [result['metadata']['text'] for result in results['matches']]

# CSV 파일을 읽고 요약하는 작업 수행
def main():
    # CSV 파일 읽기
    df = load_csv(csv_file_path)

    # 1. 텍스트 청크 나누기
    texts = []
    for index, row in df.iterrows():
        text = row['preferredExperience']  # 해당 행의 'preferredExperience' 컬럼에서 텍스트 추출
        if pd.isna(text):  # NaN 값은 건너뜁니다
            continue
        chunks = chunk_text(text)  # 텍스트를 청크로 나눕니다.
        texts.extend(chunks)  # 모든 청크를 텍스트 목록에 추가

    # 2. Pinecone DB에 텍스트 임베딩 추가
    add_to_pinecone(texts)

    # 3. 검색어를 기반으로 텍스트 찾기
    queries = ['프론트엔드', '백엔드', '풀스택']  # 검색어 목록

    for query in queries:
        print(f"### '{query}' 관련 텍스트 요약")

        # 4. Pinecone에서 검색하여 관련 텍스트 찾기
        search_results = search_pinecone(query)

        # 5. 검색된 텍스트 요약
        for text in search_results:
            summary = summarize_text(text)
            print(f"**요약:** {summary}")
            print("-" * 50)

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
from sentence_transformers import SentenceTransformer
import chromadb
import numpy as np
from transformers import pipeline

# Hugging Face의 Summarization 파이프라인 로드
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# CSV 파일 경로
csv_file_path = '/content/test_df.csv'  # 업로드된 파일 경로

# CSV 파일 읽기
def load_csv(file_path):
    return pd.read_csv(file_path)

# 텍스트 요약하는 함수
def summarize_text(text):
    if not isinstance(text, str) or not text.strip():  # 텍스트가 비어있거나 문자열이 아닐 경우
        return "내용이 없습니다."
    summary = summarizer(text, max_length=150, min_length=50, do_sample=False)
    return summary[0]['summary_text']

# 텍스트 청크로 나누기
def chunk_text(text, chunk_size=512):
    # 텍스트가 문자열인지 확인
    if not isinstance(text, str):
        text = str(text)  # 숫자나 NaN 값이 있을 경우 문자열로 변환

    # 텍스트를 청크 크기(chunk_size)로 나눕니다.
    chunks = [text[i:i + chunk_size] for i in range(0, len(text), chunk_size)]
    return chunks

# Chroma에 텍스트 임베딩 추가
def add_to_chroma(texts):
    model = SentenceTransformer('paraphrase-MiniLM-L6-v2')  # 텍스트를 임베딩할 모델
    embeddings = model.encode(texts)  # 텍스트 임베딩

    # Chroma에 임베딩 추가
    client = chromadb.Client()
    collection = client.create_collection("texts_collection")  # 텍스트를 저장할 컬렉션 생성

    # 텍스트와 임베딩을 ChromaDB에 저장
    for i, text in enumerate(texts):
        collection.add(documents=[text], embeddings=[embeddings[i]], metadatas=[{"text": text}], ids=[str(i)])

    return client, collection

# 검색어에 해당하는 텍스트 찾기
def search_chroma(collection, query, k=3):
    model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
    query_embedding = model.encode([query])  # 검색어 임베딩

    # Chroma에서 가장 가까운 k개의 벡터 찾기
    results = collection.query(query_embeddings=query_embedding, n_results=k)
    return results['documents']  # 해당하는 문서 반환

# CSV 파일을 읽고 요약하는 작업 수행
def main():
    # CSV 파일 읽기
    df = load_csv(csv_file_path)

    # 1. 텍스트 청크 나누기
    texts = []
    for index, row in df.iterrows():
        text = row['preferredExperience']  # 해당 행의 'preferredExperience' 컬럼에서 텍스트 추출
        if pd.isna(text):  # NaN 값은 건너뜁니다
            continue
        chunks = chunk_text(text)  # 텍스트를 청크로 나눕니다.
        texts.extend(chunks)  # 모든 청크를 텍스트 목록에 추가

    # 2. Chroma에 텍스트 임베딩 추가
    client, collection = add_to_chroma(texts)

    # 3. 검색어를 기반으로 텍스트 찾기
    queries = ['프론트엔드', '백엔드', '풀스택']  # 검색어 목록

    for query in queries:
        print(f"### '{query}' 관련 텍스트 요약")

        # 4. Chroma에서 검색하여 관련 텍스트 찾기
        search_results = search_chroma(collection, query)

        # 5. 검색된 텍스트 요약
        for text in search_results:
            summary = summarize_text(text)
            print(f"**요약:** {summary}")
            print("-" * 50)

if __name__ == "__main__":
    main()
```

# 문서임베딩과 유사도 측정

```python
# 문서 임베딩 모델 사용
# 예 : 자격요건, 기본 요건은 같은 말이므로 공통된 내용은 한 번에 추려서 출력
# 유사한 항목 묶고 유사도 측정
```

```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# CSV 파일 경로
csv_file_path = '/content/test_df.csv'  # 업로드된 파일 경로

# CSV 파일 읽기
def load_csv(file_path):
    return pd.read_csv(file_path)

# jobCategory 기준으로 requirement 정보를 빠르게 찾는 함수
def get_requirements_by_category(df, category):
    # jobCategory 기준으로 필터링
    category_data = df[df['jobCategory'] == category]

    # 해당 카테고리에 대한 requirement 정보만 반환 (중복 포함)
    requirements = category_data['requirement'].dropna().unique()
    return requirements

# 문장 임베딩 생성 및 유사도 계산
def get_similarity(groups):
    model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

    # 각 항목을 임베딩 벡터로 변환
    embeddings = model.encode(groups)

    # 유사도 계산 (코사인 유사도)
    similarity_matrix = cosine_similarity(embeddings)

    return similarity_matrix

# 유사도를 기반으로 그룹화된 항목 출력
def print_similar_requirements(requirements):
    similarity_matrix = get_similarity(requirements)

    # 유사도가 0.7 이상인 항목을 동일 그룹으로 처리
    threshold = 0.7
    grouped_requirements = []
    visited = set()

    for i, row in enumerate(similarity_matrix):
        if i in visited:
            continue

        # 비슷한 항목들을 하나의 그룹으로 묶기
        group = [requirements[i]]
        visited.add(i)

        for j, similarity in enumerate(row):
            if similarity >= threshold and j != i and j not in visited:
                group.append(requirements[j])
                visited.add(j)

        grouped_requirements.append(group)

    # 출력
    print("\n### 그룹화된 요구 사항 (중복 제거):")
    for group in grouped_requirements:
        print(", ".join(group))

# CSV 파일을 읽고 요구사항을 빠르게 출력하는 작업 수행
def main():
    # CSV 파일 읽기
    df = load_csv(csv_file_path)

    # jobCategory별로 빠르게 요구사항을 찾기
    print("CSV 데이터의 jobCategory 목록:")
    print(df['jobCategory'].unique())  # jobCategory 목록 확인

    # 사용자가 원하는 jobCategory 입력 받기
    category = input("원하는 jobCategory를 입력하세요: ").strip()

    # 해당 jobCategory에 맞는 requirement 출력
    if category in df['jobCategory'].values:
        requirements = get_requirements_by_category(df, category)
        print(f"\n### '{category}' 관련 요구 사항:")

        # 유사도 기반으로 요구 사항 그룹화 후 출력
        print_similar_requirements(requirements)
    else:
        print("해당 jobCategory가 존재하지 않습니다.")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# 동의어 대체용 사전 (예시)
synonym_dict = {
    '기술 스택': [
        '필요 스킬', '필수 기술', '기술 경험', '주요 기술 스택 및 도구','기술 요구 사항'
        '개발 환경', '사용 기술 스택', '기술 스택', '이런 기술을 활용해요', '개발환경은 이렇습니다'
    ],
    '스킬셋': ['soft skills', 'hard skills'],
    '자격 요건': ['필수 자격', '필수 요구 사항', '자격 조건'],
    '경험': ['경력', '경험', '숙련도'],
}

# 동의어를 대체하는 함수
def replace_synonyms(text, synonym_dict):
    for key, synonyms in synonym_dict.items():
        for synonym in synonyms:
            text = text.replace(synonym, key)  # 동의어를 기준 표현으로 변경
    return text

# CSV 파일 경로
csv_file_path = '/content/test_df.csv'  # 업로드된 파일 경로

# CSV 파일 읽기
def load_csv(file_path):
    return pd.read_csv(file_path)

# jobCategory 기준으로 requirement 정보를 빠르게 찾는 함수
def get_requirements_by_category(df, category):
    # jobCategory 기준으로 필터링
    category_data = df[df['jobCategory'] == category]

    # 해당 카테고리에 대한 requirement 정보만 반환 (중복 포함)
    requirements = category_data['requirement'].dropna().unique()

    # 동의어 대체
    requirements = [replace_synonyms(str(req), synonym_dict) for req in requirements]
    return requirements

# 문장 임베딩 생성 및 유사도 계산
def get_similarity(groups):
    model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

    # 각 항목을 임베딩 벡터로 변환
    embeddings = model.encode(groups)

    # 유사도 계산 (코사인 유사도)
    similarity_matrix = cosine_similarity(embeddings)

    return similarity_matrix

# 유사도를 기반으로 그룹화된 항목 출력
def print_similar_requirements(requirements):
    similarity_matrix = get_similarity(requirements)

    # 유사도가 0.7 이상인 항목을 동일 그룹으로 처리
    threshold = 0.7
    grouped_requirements = []
    visited = set()

    for i, row in enumerate(similarity_matrix):
        if i in visited:
            continue

        # 비슷한 항목들을 하나의 그룹으로 묶기
        group = [requirements[i]]
        visited.add(i)

        for j, similarity in enumerate(row):
            if similarity >= threshold and j != i and j not in visited:
                group.append(requirements[j])
                visited.add(j)

        grouped_requirements.append(group)

    # 출력
    print("\n### 그룹화된 요구 사항 (중복 제거):")
    for group in grouped_requirements:
        print(", ".join(group))

# CSV 파일을 읽고 요구사항을 빠르게 출력하는 작업 수행
def main():
    # CSV 파일 읽기
    df = load_csv(csv_file_path)

    # jobCategory별로 빠르게 요구사항을 찾기
    print("CSV 데이터의 jobCategory 목록:")
    print(df['jobCategory'].unique())  # jobCategory 목록 확인

    # 사용자가 원하는 jobCategory 입력 받기
    category = input("원하는 jobCategory를 입력하세요: ").strip()

    # 해당 jobCategory에 맞는 requirement 출력
    if category in df['jobCategory'].values:
        requirements = get_requirements_by_category(df, category)
        print(f"\n### '{category}' 관련 요구 사항:")

        # 유사도 기반으로 요구 사항 그룹화 후 출력
        print_similar_requirements(requirements)
    else:
        print("해당 jobCategory가 존재하지 않습니다.")

if __name__ == "__main__":
    main()
```

```python
# 유사도 행렬
# CountVectorizer로 얻은 벡터의 크기와 유사도 행렬의 크기를 일치시키는 방법 - 함수 2개 사용
# extract_keywords 함수: CountVectorizer에서 feature_names와 함께 X를 반환하여 X는 키워드들 간의 벡터 표현을 포함하고, 이 벡터를 기반으로 유사도 계산을 진행할 수 있게 수정했습니다.
# group_similar_keywords 함수: X를 사용하여 유사도 행렬을 계산하고, 이를 바탕으로 유사한 키워드를 그룹화합니다.
# =>
# requirement 열에 포함된 텍스트에서 추출한 키워드들을 유사도 기반으로 그룹화하여 동의어 대체 사전을 생성
```

```python
import pandas as pd
import re
from collections import defaultdict
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# CSV 파일 경로
csv_file_path = '/content/test_df.csv'  # 업로드된 파일 경로

# CSV 파일 읽기
def load_csv(file_path):
    return pd.read_csv(file_path)

# 텍스트 정제 (불필요한 기호 및 공백 제거)
def clean_text(text):
    # 소문자 변환 및 특수문자 제거
    text = re.sub(r'[^가-힣a-zA-Z0-9\s]', '', text)
    text = text.lower()
    return text

# 기술 및 키워드 추출 (단순화된 방법으로 단어 추출)
def extract_keywords(texts):
    vectorizer = CountVectorizer(stop_words='english', max_features=50)  # 상위 50개 단어 추출
    X = vectorizer.fit_transform(texts)
    feature_names = vectorizer.get_feature_names_out()
    return feature_names, X

# 유사도 계산을 통한 동의어 그룹화
def group_similar_keywords(X, num_keywords):
    similarity_threshold = 0.7  # 유사도 기준
    similarity_groups = defaultdict(list)

    # 키워드 간 코사인 유사도 계산
    cosine_sim_matrix = cosine_similarity(X)

    # 유사한 키워드 그룹화
    for i, row in enumerate(cosine_sim_matrix):
        for j, similarity in enumerate(row):
            if i != j and similarity >= similarity_threshold:
                if i < num_keywords and j < num_keywords:
                    similarity_groups[i].append(j)

    return similarity_groups

# 동의어 사전 생성
def create_synonym_dict(keywords, similarity_groups):
    synonym_dict = {}
    for group in similarity_groups.values():
        group_keywords = [keywords[i] for i in group]
        key = group_keywords[0]  # 그룹의 첫 번째 키워드를 기준으로 사용
        synonym_dict[key] = group_keywords
    return synonym_dict

# 주어진 CSV 파일에서 요구 사항 분석 및 동의어 사전 생성
def create_synonym_dictionary_from_csv(file_path):
    # CSV 파일 읽기
    df = load_csv(file_path)

    # 'requirement' 열에서 텍스트 추출 및 정제
    cleaned_texts = df['requirement'].dropna().apply(clean_text)

    # 텍스트에서 기술/키워드 추출
    keywords, X = extract_keywords(cleaned_texts)
    num_keywords = len(keywords)

    # 유사도 계산을 통해 키워드 그룹화
    similarity_groups = group_similar_keywords(X, num_keywords)

    # 동의어 사전 생성
    synonym_dict = create_synonym_dict(keywords, similarity_groups)

    return synonym_dict

# 동의어 사전 생성 및 출력
def main():
    synonym_dict = create_synonym_dictionary_from_csv(csv_file_path)

    print("생성된 동의어 대체 사전:")
    for key, synonyms in synonym_dict.items():
        print(f"{key}: {', '.join(synonyms)}")

if __name__ == "__main__":
    main()
```
